```python
# Importar as bibliotecas necessárias
!pip install pandas matplotlib seaborn numpy
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configurar estilo de gráficos
sns.set(style="whitegrid")

# Carregar o dataset
file_path = 'tiktok_datasetnovo.csv'  # Substitua pelo caminho do arquivo
df = pd.read_csv(file_path)

# Exibir as primeiras linhas do dataset para inspeção
df.head()
```

```python
# Limpeza de dados: remoção de valores ausentes
df = df.dropna()

# Criação de novas métricas
df['likes_per_view'] = df['video_like_count'] / df['video_view_count']
df['comments_per_view'] = df['video_comment_count'] / df['video_view_count']

# Exibir a descrição estatística para verificar as métricas adicionadas
df.describe()
```

```python
# Plotar a distribuição de visualizações
plt.figure(figsize=(10, 6))
sns.histplot(df['video_view_count'], bins=50, kde=True, color='blue')
plt.title('Distribuição de Visualizações dos Vídeos')
plt.xlabel('Número de Visualizações')
plt.ylabel('Frequência')
plt.show()

# Plotar a distribuição de curtidas
plt.figure(figsize=(10, 6))
sns.histplot(df['video_like_count'], bins=50, kde=True, color='green')
plt.title('Distribuição de Curtidas dos Vídeos')
plt.xlabel('Número de Curtidas')
plt.ylabel('Frequência')
plt.show()
```

```python
# Scatter plot: visualizações vs. curtidas
plt.figure(figsize=(10, 6))
sns.scatterplot(x='video_view_count', y='video_like_count', data=df, alpha=0.6)
plt.title('Relação entre Visualizações e Curtidas')
plt.xlabel('Número de Visualizações')
plt.ylabel('Número de Curtidas')
plt.show()

# Scatter plot: curtidas vs. compartilhamentos
plt.figure(figsize=(10, 6))
sns.scatterplot(x='video_like_count', y='video_share_count', data=df, alpha=0.6, color='orange')
plt.title('Relação entre Curtidas e Compartilhamentos')
plt.xlabel('Número de Curtidas')
plt.ylabel('Número de Compartilhamentos')
plt.show()
```

```python
# Calcular a matriz de correlação
corr = df[['video_view_count', 'video_like_count', 'video_comment_count',
           'video_duration_sec', 'video_download_count', 'video_share_count',
           'likes_per_view', 'comments_per_view']].corr()

# Plotar o mapa de correlação
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='Blues', fmt=".2f")
plt.title('Mapa de Correlação Entre Métricas')
plt.show()
```

```python
# Selecionar os 10 vídeos mais populares
top_videos = df.sort_values(by='video_view_count', ascending=False).head(10)

# Exibir os dados dos Top 10 vídeos
print(top_videos[['video_id', 'video_view_count', 'video_like_count', 'video_comment_count', 'video_duration_sec']])

# Gráfico de barras: visualizações dos Top 10 vídeos
plt.figure(figsize=(12, 6))
sns.barplot(x='video_view_count', y='video_id', data=top_videos, palette='viridis')
plt.title('Top 10 Vídeos Mais Populares (Visualizações)')
plt.xlabel('Número de Visualizações')
plt.ylabel('ID do Vídeo')
plt.show()
```

```python
# Simulação: aumento de curtidas e impacto nos compartilhamentos
simulated_data = df[['video_like_count', 'video_share_count']].copy()
simulated_data['likes_increase'] = simulated_data['video_like_count'] * 1.1
simulated_data['predicted_shares'] = simulated_data['likes_increase'] * corr.loc['video_like_count', 'video_share_count']

# Gráfico do impacto esperado
plt.figure(figsize=(10, 6))
sns.scatterplot(x='video_like_count', y='video_share_count', data=simulated_data, label='Atual')
sns.scatterplot(x='likes_increase', y='predicted_shares', data=simulated_data, label='Impacto Esperado', color='orange')
plt.title('Impacto de Aumentar Curtidas nos Compartilhamentos')
plt.xlabel('Curtidas')
plt.ylabel('Compartilhamentos')
plt.legend()
plt.show()
```

```python

# Criar métricas adicionais
df['likes_per_view'] = df['video_like_count'] / df['video_view_count']
df['comments_per_view'] = df['video_comment_count'] / df['video_view_count']

# Definir variáveis independentes (X) e dependente (Y)
x_cols = ['video_like_count', 'video_comment_count', 'video_duration_sec']  # Ajuste as colunas conforme necessário
y_col = 'video_view_count'

# Criar os dados de X e Y
X = df[x_cols]
y = df[y_col]

# Verificar tipos e converter para numérico
X = X.apply(pd.to_numeric, errors='coerce')
y = pd.to_numeric(y, errors='coerce')

# Remover valores ausentes
X = X.dropna()
y = y.dropna()

# Adicionar uma constante ao modelo (para o intercepto)
X = sm.add_constant(X)

# Verificar variância das colunas e remover constantes
print("Variância das colunas antes do ajuste:")
print(X.var())
X = X.loc[:, X.var() > 0]

# Ajustar o modelo de regressão linear
try:
    model = sm.OLS(y, X).fit()
    print(model.summary())
except Exception as e:
    print(f"Erro ao ajustar o modelo: {e}")
```
